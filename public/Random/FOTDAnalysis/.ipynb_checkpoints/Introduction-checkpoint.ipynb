```python
import json # process json files
import matplotlib # plot results
import matplotlib.pyplot as plt
import glob, os # get chat files
%matplotlib inline
plt.style.use('dark_background')
```

First, we want to find all the JSON files that are in the channels folder, we're going to use a built-in module called glob and os to do so.

```python
files, names = [], []
for file in glob.glob("channels/*"):
    names.append(file[9:-5]) # all the files end with .json so indexing at [:-5] will give all the characters except the last 5
print(f"Names:\t {names}")
```

Now that we have a list of important files, I want to create a dictionary that maps from name to a list of message objects (We'll look at what a message object is later)

```python
message_data = {}

for n in names:
    messages = json.loads(open(f"channels/{n}.json", 'r').read())["messages"]
    message_data[n] = messages
```

Now that we have all the messages, let's take a look at what one message object means

```python
message_data["jonathan"][0]
```

It contains all the information about the message that you could possible get from looking at it, including reactions. Now we can start doing some analysis of all the FOTD channels.

Let's start off with who has posted the most

```python
msg_counts = [(len(message_data[n]), n) for n in names]
msg_counts.sort(key=lambda a: -a[0])
print(msg_counts)
```

```python
fig, ax = plt.subplots(figsize=(22, 10))

_=ax.bar([x[1] for x in msg_counts], [x[0] for x in msg_counts])

ax.set_ylabel("Messages Sent")
ax.set_title("FOTD Channel Messages Sent");

fig.savefig("total_messages.png")
```

A congratulations to Tompson for leading the charts in total activity... but some people joined the FOTD bandwagon late, the statisitic we really want is <b>messages per day</b>

This is a little more tricky since we need to get the date the first message was sent and do some math but it's still managable

```python
import datetime

channel_lifespan = {}
for n in names:
    channel_lifespan[n] = (datetime.datetime.now() -\
    datetime.datetime.strptime(message_data[n][0]["timestamp"], "%Y-%m-%dT%H:%M:%S.%f+00:00")).days
print(channel_lifespan)
```

```python
msg_per_day = [(x[0]/channel_lifespan[x[1]], x[1]) for x in msg_counts]
msg_per_day.sort(key=lambda x: -x[0])

fig, ax = plt.subplots(figsize=(22, 10))
_=ax.bar([x[1] for x in msg_per_day], [x[0] for x in msg_per_day])

ax.set_ylabel("Messages Sent")
ax.set_title("FOTD Channel Messages Sent");
ax.set_xlabel("Channel");

fig.savefig("messages_per_day.png")
```
